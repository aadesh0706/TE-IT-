```python
cars_data = pd.read_csv('DS1_C5_S1_Descriptive_Analysis_Practice_Car_Data.csv')
cars_data
```

```python
s_car = cars_data[['Make','Model','Displacement','City_Mileage_km_litre','Highway_Mileage_km_litre','Fuel_Tank_Capacity_litre','Minimum_Turning_Radius']]
s_car
```

```python
s_car.isnull().sum()
```

```python
s_car.dropna(inplace=True)
```

```python
s_car
```

```python
s_car.describe()
```

```python
demo3 = s_car.describe().loc[['mean','50%']]
demo3
```

```python
demo3.loc['S.D.',:] = s_car.std()
demo3
```

```python
demo3.loc['Coeff.Var.',:] = (s_car.std()/s_car.mean())*100
demo3
```
